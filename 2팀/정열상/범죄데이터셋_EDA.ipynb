```python
import pandas as pd
import numpy as np
from sklearn.neighbors import BallTree
from joblib import Parallel, delayed

# 데이터 전처리
df['Date'] = pd.to_datetime(df['Date'])
df.dropna(subset=['X Coordinate', 'Y Coordinate'], inplace=True)

# 2012년~2016년 데이터만 필터링
df = df[(df['Date'].dt.year >= 2012) & (df['Date'].dt.year <= 2016)]

# 각 Beat에서 가장 빈번한 상위 3가지 범죄 선정
top3_crimes_by_beat = (
    df.groupby(['Beat', 'Primary Type']).size()
    .reset_index(name='Count')
    .sort_values(['Beat', 'Count'], ascending=[True, False])
    .groupby('Beat').head(3)
)

# 주요 범죄 데이터만 필터링
filtered_df = df.merge(top3_crimes_by_beat[['Beat', 'Primary Type']], on=['Beat', 'Primary Type'], how='inner').reset_index()

# BallTree 기반 공간 검색 준비 (X/Y 좌표 사용)
coordinates = df[['X Coordinate', 'Y Coordinate']].values
tree = BallTree(coordinates, metric='euclidean')

# 동일 범죄 재발 시점 탐색 함수 정의
def find_next_same_crime(tree, df, base_idx, radius=100):
    base_crime = df.iloc[base_idx]
    base_date = base_crime['Date']
    crime_type = base_crime['Primary Type']
    base_point = coordinates[base_idx].reshape(1, -1)

    # base 이후 발생한 같은 유형의 사건만 필터링
    mask = (df['Date'] > base_date) & (df['Primary Type'] == crime_type)
    candidate_indices = df[mask].index

    # 반경 내 후보 추출
    indices_within_radius = tree.query_radius(base_point, r=radius)[0]
    nearby_same_crimes = list(set(indices_within_radius).intersection(set(candidate_indices)))

    if not nearby_same_crimes:
        return {
            'Beat': base_crime['Beat'],
            'Primary Type': crime_type,
            'Crime Date': base_date,
            'Days_to_next_same_type': np.nan,
            'Next Crime Date': pd.NaT
        }

    # 가장 빠른 날짜 찾기
    next_dates = df.loc[nearby_same_crimes, 'Date']
    next_crime_date = next_dates.min()
    days_diff = (next_crime_date - base_date).days

    return {
        'Beat': base_crime['Beat'],
        'Primary Type': crime_type,
        'Crime Date': base_date,
        'Days_to_next_same_type': days_diff,
        'Next Crime Date': next_crime_date
    }

# 병렬 처리 실행
results = Parallel(n_jobs=-1, backend='multiprocessing')(
    delayed(find_next_same_crime)(tree, df, idx) for idx in filtered_df.index
)

# 결과를 데이터프레임으로 정리
results_df = pd.DataFrame(results)

# 결과 저장
results_df.to_csv('same_crime_recurrence_results.csv', index=False)

# 결과 확인
print(results_df.head())
```
