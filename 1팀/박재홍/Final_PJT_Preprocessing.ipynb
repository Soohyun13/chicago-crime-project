```python
# 모듈 불러오기
import pandas as pd
import numpy as np
```

```python
# 데이터 불러오기
df = pd.read_csv("data/merge_data.csv", low_memory=False)
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 각 Block 별 Latitude, Longitude의 최빈값(Mode)으로 결측치 채우기
df['Latitude'] = df.groupby('Block')['Latitude'].transform(lambda x: x.mode()[0] if not x.mode().empty else x)
df['Longitude'] = df.groupby('Block')['Longitude'].transform(lambda x: x.mode()[0] if not x.mode().empty else x)

# Block 별로 Location의 최빈값을 구하고 결측치를 채우기
df['Location'] = df.groupby('Block')['Location'].transform(lambda x: x.mode()[0] if not x.mode().empty else x)

# Block별로 X Coordinate와 Y Coordinate의 최빈값을 사용하여 결측치 채우기
df['X Coordinate'] = df.groupby('Block')['X Coordinate'].transform(lambda x: x.mode()[0] if not x.mode().empty else x.mean())
df['Y Coordinate'] = df.groupby('Block')['Y Coordinate'].transform(lambda x: x.mode()[0] if not x.mode().empty else x.mean())
```

```python
# 각 community_area 별로 Latitude, Longitude의 최빈값(Mode) 계산
mode_values = df.groupby('Community Area')[['Latitude', 'Longitude', 'Location', 'X Coordinate', 'Y Coordinate']].agg(lambda x: x.mode()[0])

# 결측치를 각 Community Area 별로 최빈값으로 채우기
for community in mode_values.index:
    df.loc[(df['Community Area'] == community) & (df['Latitude'].isna()), 'Latitude'] = mode_values.loc[community, 'Latitude']
    df.loc[(df['Community Area'] == community) & (df['Longitude'].isna()), 'Longitude'] = mode_values.loc[community, 'Longitude']
    df.loc[(df['Community Area'] == community) & (df['Location'].isna()), 'Location'] = mode_values.loc[community, 'Location']
    df.loc[(df['Community Area'] == community) & (df['X Coordinate'].isna()), 'X Coordinate'] = mode_values.loc[community, 'X Coordinate']
    df.loc[(df['Community Area'] == community) & (df['Y Coordinate'].isna()), 'Y Coordinate'] = mode_values.loc[community, 'Y Coordinate']

# 결측치가 채워졌는지 확인
print(df[['Community Area', 'Latitude', 'Longitude', 'Location','X Coordinate', 'Y Coordinate']].isna().sum())
```

```python
# 각 community_area 별로 Latitude, Longitude의 최빈값(Mode) 계산
mode_values = df.groupby('Ward')[['Latitude', 'Longitude', 'Location', 'X Coordinate', 'Y Coordinate']].agg(lambda x: x.mode()[0])

# 결측치를 각 Community Area 별로 최빈값으로 채우기
for community in mode_values.index:
    df.loc[(df['Ward'] == community) & (df['Latitude'].isna()), 'Latitude'] = mode_values.loc[community, 'Latitude']
    df.loc[(df['Ward'] == community) & (df['Longitude'].isna()), 'Longitude'] = mode_values.loc[community, 'Longitude']
    df.loc[(df['Ward'] == community) & (df['Location'].isna()), 'Location'] = mode_values.loc[community, 'Location']
    df.loc[(df['Ward'] == community) & (df['X Coordinate'].isna()), 'X Coordinate'] = mode_values.loc[community, 'X Coordinate']
    df.loc[(df['Ward'] == community) & (df['Y Coordinate'].isna()), 'Y Coordinate'] = mode_values.loc[community, 'Y Coordinate']

# 결측치가 채워졌는지 확인
print(df[['Ward', 'Latitude', 'Longitude', 'Location','X Coordinate', 'Y Coordinate']].isna().sum())
```

```python
import geopandas as gpd

# 시카고 Ward GeoJSON 불러오기 (구글 드라이브 경로)
gdf_wards = gpd.read_file("data/chicago-wards.geojson") # 맞게 변환

# 좌표계 확인 및 설정
if gdf_wards.crs is None:
    gdf_wards.set_crs(epsg=4326, inplace=True)
else:
    gdf_wards = gdf_wards.to_crs(epsg=4326)

import geopandas as gpd
from shapely.geometry import Point

# 꼭 필요한 열만 추출
df_coords = df.loc[
    df['Latitude'].notnull() & df['Longitude'].notnull() & df['Ward'].isnull(),
    ['Latitude', 'Longitude']
].copy()

# 벡터 연산으로 Point 생성
geometry = gpd.points_from_xy(df_coords['Longitude'].astype(float), df_coords['Latitude'].astype(float))

# GeoDataFrame 생성
gdf_coords = gpd.GeoDataFrame(df_coords, geometry=geometry, crs='EPSG:4326')

gdf_wards = gpd.read_file("data/chicago-wards.geojson")
gdf_wards = gdf_wards.to_crs(epsg=4326)

# Spatial join
joined = gpd.sjoin(gdf_coords, gdf_wards, how='left', predicate='within')

# ward 컬럼명 자동 감지
ward_col_name = [col for col in joined.columns if 'ward' in col.lower()][0]

# 인덱스 정렬 맞춰서 채우기
df.loc[joined.index, 'Ward'] = joined[ward_col_name].values
```

```python
# 시카고 커뮤니티 지역 GeoJSON 로드
gdf_community = gpd.read_file("data/chicago-community-areas.geojson") # 맞게 수정
gdf_community = gdf_community.to_crs(epsg=4326)

import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# 커뮤니티 지역 GeoJSON 로드
gdf_community = gpd.read_file("data/chicago-community-areas.geojson") # 맞게 수정
gdf_community = gdf_community.to_crs(epsg=4326)

# 결측치 있는 위경도 데이터만 필터링
df_missing_comm = df.loc[
    df['Community Area'].isnull() &
    df['Latitude'].notnull() &
    df['Longitude'].notnull(),
    ['Latitude', 'Longitude']
].copy()

# 인덱스 보존
df_missing_comm['orig_index'] = df_missing_comm.index

# 💡 데이터가 많으니 5000개 단위로 나눠서 Spatial Join
batch_size = 5000
filled_area = []

for i in range(0, len(df_missing_comm), batch_size):
    batch = df_missing_comm.iloc[i:i+batch_size].copy()
    # Point 객체 생성 (벡터화)
    geometry = gpd.points_from_xy(batch['Longitude'].astype(float), batch['Latitude'].astype(float))
    gdf_batch = gpd.GeoDataFrame(batch, geometry=geometry, crs='EPSG:4326')

    # Spatial Join
    joined = gpd.sjoin(gdf_batch, gdf_community, how='left', predicate='within')

    # community area 컬럼명 자동 탐색
    comm_col = [col for col in joined.columns if 'community' in col.lower() or 'area' in col.lower()][0]

    # 결과 저장
    filled_area.append(joined[['orig_index', comm_col]])

# 결과 통합
area_result = pd.concat(filled_area)

# 원본 df에 반영
df.loc[area_result['orig_index'], 'Community Area'] = area_result[comm_col].values
```
