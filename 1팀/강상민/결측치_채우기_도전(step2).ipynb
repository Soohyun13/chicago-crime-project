- ID: 해당 기록의 고유 식별자.
- Case Number: 사건의 고유 번호(시카고 경찰서 기록부 번호, RD 번호).
- Date: 사건이 발생한 날짜.
- Block: 사건이 발생한 주소(블록 단위).
- IUCR: 일리노이 범죄 보고 코드(Illinois Uniform Crime Reporting Code).
- Primary Type: IUCR 코드의 주요 범죄 유형.
- Description: IUCR 코드의 세부 범죄 유형(Primary Type의 하위 범주).
- Location Description: 사건이 발생한 장소의 설명.
- Arrest: 체포 여부.
- Domestic: 해당 사건이 일리노이 가정 폭력법에 따라 가정폭력 관련 사건인지 여부.
- Beat: 사건이 발생한 경찰 순찰 구역(가장 작은 경찰 행정 단위).
- District: 사건이 발생한 경찰서 관할 구역.
- Ward: 사건이 발생한 시의회 선거구(워드).
- Community Area: 사건이 발생한 커뮤니티 지역(시카고에는 총 77개의 커뮤니티 지역이 있음).
- FBI Code: FBI 국가 사건 기반 보고 시스템(NIBRS) 기준의 범죄 분류 코드.
- X Coordinate: 사건이 발생한 위치의 X 좌표(State Plane Illinois East NAD 1983 기준).
- Y Coordinate: 사건이 발생한 위치의 Y 좌표(State Plane Illinois East NAD 1983 기준).
- Year: 사건이 발생한 연도.
- Updated On: 해당 기록이 마지막으로 업데이트된 날짜 및 시간.
- Latitude: 사건이 발생한 위치의 위도(개인 정보 보호를 위해 실제 위치에서 약간 이동됨).
- Longitude: 사건이 발생한 위치의 경도(개인 정보 보호를 위해 실제 위치에서 약간 이동됨).
- Location: 사건이 발생한 위치를 지도 제작 및 기타 지리적 분석이 가능하도록 저장한 좌표(개인 정보 보호를 위해 실제 위치에서 약간 이동됨).

```python
df_s = df.copy()
df_s = df_s.sort_values(by='Block')
df_s
```

```python
# 결측치 채우기
# 결측치 없는 유일한 주소 정보인 block을 기준으로 결측치를 채우기
# 평균으로 하면 잘못 적은 데이터에 의해 정보가 왜곡될 수 있으므로 최빈값 사용

df_s['Latitude'] = df_s['Latitude'].fillna(df_s.groupby('Block')['Latitude'].transform(lambda x: x.mode()[0] if not x.mode().empty else x))
df_s['Longitude'] = df_s['Longitude'].fillna(df_s.groupby('Block')['Longitude'].transform(lambda x: x.mode()[0] if not x.mode().empty else x))

df_s['Ward'] = df_s['Ward'].fillna(df_s.groupby('Block')['Ward'].transform(lambda x: x.mode()[0] if not x.mode().empty else x))
df_s['Community Area'] = df_s['Community Area'].fillna(df_s.groupby('Block')['Community Area'].transform(lambda x: x.mode()[0] if not x.mode().empty else x))

df_s['X Coordinate'] = df_s['X Coordinate'].fillna(df_s.groupby('Block')['X Coordinate'].transform(lambda x: x.mode()[0] if not x.mode().empty else x))
df_s['Y Coordinate'] = df_s['Y Coordinate'].fillna(df_s.groupby('Block')['Y Coordinate'].transform(lambda x: x.mode()[0] if not x.mode().empty else x))

df_s['Location'] = df_s['Location'].fillna('( '+ df_s['Latitude'].astype('str') + ', ' + df_s['Longitude'].astype('str') + ')')

df_s.isna().sum()
```

```python
import geopandas as gpd
from shapely.geometry import Point

wards_gdf = gpd.read_file("Boundaries - Wards (2023-)_20250328.geojson")

df_s["Latitude"] = df_s["Latitude"].astype(float)
df_s["Longitude"] = df_s["Longitude"].astype(float)

# crime_df = pd.read_csv("chicago_crime_data.csv")

df_s["geometry"] = df_s.apply(lambda row: Point(row["Longitude"], row["Latitude"]), axis=1)

crime_gdf = gpd.GeoDataFrame(df_s, geometry="geometry", crs="EPSG:4326")

crime_gdf = gpd.sjoin(crime_gdf, wards_gdf[["ward", "geometry"]], how="left", predicate="within")

df_s["Ward"].fillna(crime_gdf["ward"], inplace=True)

df_s.info()
```

```python
# cond = (df_s['Ward'] == np.nan)

# df_s[cond].shape

df_s['Ward'] = df_s['Ward'].astype('float')
df_s.info()
```

```python
geo_data = gpd.read_file("CommAreas_20250328.geojson")

df_s["geometry"] = df_s.apply(lambda row: Point(row["Longitude"], row["Latitude"]), axis=1)

gdf = gpd.GeoDataFrame(df_s, geometry='geometry', crs="EPSG:4326")

gdf = gdf.sjoin(geo_data[['geometry', 'area_num_1']], how="left", predicate="within")

df_s['Community Area'] = df_s['Community Area'].fillna(gdf['area_num_1'])
```
