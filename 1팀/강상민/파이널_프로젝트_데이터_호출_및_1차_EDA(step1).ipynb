- ID: 해당 기록의 고유 식별자.
- Case Number: 사건의 고유 번호(시카고 경찰서 기록부 번호, RD 번호).
- Date: 사건이 발생한 날짜.
- Block: 사건이 발생한 주소(블록 단위).
- IUCR: 일리노이 범죄 보고 코드(Illinois Uniform Crime Reporting Code).
- Primary Type: IUCR 코드의 주요 범죄 유형.
- Description: IUCR 코드의 세부 범죄 유형(Primary Type의 하위 범주).
- Location Description: 사건이 발생한 장소의 설명.
- Arrest: 체포 여부.
- Domestic: 해당 사건이 일리노이 가정 폭력법에 따라 가정폭력 관련 사건인지 여부.
- Beat: 사건이 발생한 경찰 순찰 구역(가장 작은 경찰 행정 단위).
- District: 사건이 발생한 경찰서 관할 구역.
- Ward: 사건이 발생한 시의회 선거구(워드).
- Community Area: 사건이 발생한 커뮤니티 지역(시카고에는 총 77개의 커뮤니티 지역이 있음).
- FBI Code: FBI 국가 사건 기반 보고 시스템(NIBRS) 기준의 범죄 분류 코드.
- X Coordinate: 사건이 발생한 위치의 X 좌표(State Plane Illinois East NAD 1983 기준).
- Y Coordinate: 사건이 발생한 위치의 Y 좌표(State Plane Illinois East NAD 1983 기준).
- Year: 사건이 발생한 연도.
- Updated On: 해당 기록이 마지막으로 업데이트된 날짜 및 시간.
- Latitude: 사건이 발생한 위치의 위도(개인 정보 보호를 위해 실제 위치에서 약간 이동됨).
- Longitude: 사건이 발생한 위치의 경도(개인 정보 보호를 위해 실제 위치에서 약간 이동됨).
- Location: 사건이 발생한 위치를 지도 제작 및 기타 지리적 분석이 가능하도록 저장한 좌표(개인 정보 보호를 위해 실제 위치에서 약간 이동됨).

### beat를 기준으로 위치 결측을 채워보려 했더니
- 위경도와 beat 사이의 상관관계가 낮아 일정한 기준을 가지고 위도와 경도를 구간화하기 어려움
- 실제로 시카고의 beat는 가장 낮은 번호가 남동쪽에 있고 특별한 방향 기준 없이 beat의 고유 번호가 나열되음

image.png

 - 규모가 큰 beat는 경찰서의 좌표 값으로 대입하면 실제 범죄 발생 현장과 위치 차이가 클 수 있음 → 경찰서 위치 기준으로 결측치를 채우는 데는 한계가 뚜렷

∴ 결론 : 결측치를 온전하게 채우기 힘드니 버리는 것이 낫다
