```python
# 시카고 커뮤니티 area 의 실제 개수는 77
# 0이나 77이 누락 데이터인지 확인
print(len(chicago[chicago['Community Area'] == 77]))
print(len(chicago[chicago['Community Area'] == 0]))
```

### CM 0번 제거 version

- 95개로 0.000012%의 작은 데이터 이므로 제거

```python
# 'Community Area'가 0인 행 제거
chicago = chicago[chicago['Community Area'] != 0]
```

```python
area_name = area_name.iloc[1:,0:2]
```

```python
area_name.head()
```

```python
area_name = area_name.rename(columns={'name' : 'ca_name'})
```

### chicago 데이터에 이름 결합

```python
chicago['Community Area'] = chicago['Community Area'].astype(int)
```

```python
# 'Community Area' 컬럼을 모두 문자열로 변환
chicago['Community Area'] = chicago['Community Area'].astype(str)
area_name['Community Area'] = area_name['Community Area'].astype(str)
```

```python
chicago['Community Area'].dtype
```

```python
chicago['Community Area']
```

```python
merge = pd.merge(chicago, area_name, how='left', on='Community Area')
```
